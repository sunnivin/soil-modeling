```python
from sympy import symbols, plot_implicit, Eq, Abs, solve 
from matplotlib import style
style.use('ggplot')
p, q = symbols('p q')

max_p = 150 
# Ellipse 1 
a1=max_p/2  # Major axis 
b1=max_p/2  # Minor axis 
h1=a1  # Coordinate of ellipse center along x-axis
k1=0    # Coordinate of ellipse center along y-axis 
ellipse1 = Eq(((p-h1)**2)/a1**2 + ((q-k1)**2)/b1**2, 1)

# Line 
x_slope = M
line=Eq(p * x_slope + 0,q)

# Stress path 
x_slope = q_over_p 

# calculate b 
p_start = 100
b = -x_slope*p_start
stress_path=Eq(p*x_slope+b,q)


plot1 = plot_implicit(ellipse1, (p, -0.1, 155), (q, -0.1, 160),xlabel="p'",ylabel='q',
                      line_color='steelblue', show=False)
plot2 = plot_implicit(line, (p, -0.1, 155), (q, -0.1, 160),
                      line_color='crimson', show=False)
plot3 = plot_implicit(stress_path,(p,-0.1,155),(q,-0.1,160),line_color='black', show=False)


# Append to show in same graph
plot1.append(plot2[0])
plot1.append(plot3[0])
plot1.show()


# Find crossing points between line and ellipse1 
crossing_point = [ellipse1,line]
answer = solve(crossing_point,(p,q))
print(f"Crossing point with line and ellipse1 {answer}")

# Find crossing points between stress path and ellipse1 
crossing_point = [ellipse1,stress_path]
answer = solve(crossing_point,(p,q))
print(f"Crossing point between stress path and ellipse2 {answer}")
```

```python
from sympy import symbols, plot_implicit, Eq, Abs, solve 
from matplotlib import style
style.use('ggplot')
p, q, delta_p, delta_q = symbols('p q delta_p delta_q')


q = 0 
delta_q = 3.2*delta_p
M=1.2 
p_m = 150 
p = 100


yield_surface = (q+delta_q)**2-M**2*((p+delta_p)*(p_m-(p+delta_p)))

answer = solve(yield_surface,delta_p)

print(f"delta_p {answer}")

positive_delta_p = answer[1]
print(f"positive_delta_p {positive_delta_p}")
```

```python
# updated q/p ratio at yield 
p = 100
delta_q = 3.2*positive_delta_p
p2 = p + positive_delta_p 
q2 = 0. + delta_q

q_over_p = q2/p2

print(f"updated q_over_p ratio : {q_over_p}")
```

```python

K_0 = (3-q_over_p)/((2*q_over_p)+3)



print(f"K_0 {K_0}")
```
